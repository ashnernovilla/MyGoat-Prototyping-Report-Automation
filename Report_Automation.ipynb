```python
# @title Goat Total Cuts
Goat_Utilization_cuts = pd.concat([Goat_Utilization_Data_1, Goat_Utilization_Data_2], axis=0)
Goat_Utilization_cuts['status_timestamp'] = pd.to_datetime(Goat_Utilization_cuts['status_timestamp'])

# Sort by status_timestamp
Goat_Utilization_cuts = Goat_Utilization_cuts.sort_values(by='status_timestamp').reset_index(drop=True)

# Create YearWeekTag column
Goat_Utilization_cuts['YearWeekTag'] = Goat_Utilization_cuts['status_timestamp'].apply(
    lambda x: f"Year {x.year} - Week {x.isocalendar().week:02d}"
)

# display(Goat_Utilization_cuts.head())

# Filter to MOWING activity
mowing_df = Goat_Utilization_cuts[Goat_Utilization_cuts['activity'] == "MOWING"]

# Count of unique weeks
countofweeks = mowing_df['YearWeekTag'].nunique()

# Total hours = count of rows * 15 min ÷ 60
sumofhours_season = len(mowing_df) * 15 / 60

# Average hours cut per week
average_hours_cut_per_week = sumofhours_season / countofweeks if countofweeks > 0 else 0

# display(mowing_df.head())

# Step 1: Create weekly mowing stats per goat or customer
goat_weekly = (
    mowing_df
    .groupby(['goat_id', 'goat_name', 'customer_name', 'YearWeekTag'])
    .agg(mowing_records=('goat_id', 'count'))  # or use 'status_timestamp'
    .reset_index()
)

# Step 2: Calculate average hours per week per goat
# 15 minutes per row = 0.25 hours
goat_summary = (
    goat_weekly
    .groupby(['goat_id', 'goat_name', 'customer_name'])
    .agg(
        weeks_cut=('YearWeekTag', 'nunique'),
        total_hours=('mowing_records', lambda x: x.sum() * 0.25)
    )
    .reset_index()
)

goat_summary['AverageHoursCutPerWeek'] = goat_summary['total_hours'] / goat_summary['weeks_cut']

# Step 3: Define conditional cut multipliers based on customer
conditions = [
    goat_summary['customer_name'] == "Nashville Metro Dell Soccer Complex",
    goat_summary['customer_name'] == "City of Birmingham - Harrison Park",
    goat_summary['customer_name'] == "Fall Line Commercial Services",
    goat_summary['customer_name'] == "Lipscomb Academy",
    goat_summary['customer_name'] == "Savannah Hilton Head Airport",
]

multipliers = [5000, 3000, 3000, 5000, 5000]
default_multiplier = 3000

goat_summary['SqftPerHour'] = np.select(conditions, multipliers, default=default_multiplier)


print("Average Mowing Hours For all the Goat in a Week: ", average_hours_cut_per_week, "Hours")

average_sqft_cut_per_week = (average_hours_cut_per_week * goat_summary.loc[goat_summary.index == 0, 'SqftPerHour'].values[0])/42000
average_sqft_cut_per_week = round(average_sqft_cut_per_week, 0)

print("Average Mowing Sqft Cut For all the Goat in a Week: ", average_sqft_cut_per_week, "sqft")
```
